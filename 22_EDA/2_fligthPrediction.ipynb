```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('business.csv')
df.head()
```

```python
## get the basic info about the data 
df['price'] = df['price'].str.replace(',', '').astype(int)
df.info()
```

```python
df.describe()
```

```python
## Featrure Engineering
df['Date of Journey'] = df['date'].str.split('-').str[0]
df['Month of Journey'] = df['date'].str.split('-').str[1]
df['Year of Journey'] = df['date'].str.split('-').str[2]
```

```python
df.head()
```

```python
df['Date of Journey'] = df['Date of Journey'].astype(int)
df['Month of Journey'] = df['Month of Journey'].astype(int)
df['Year of Journey'] = df['Year of Journey'].astype(int)
```

```python
df.info()
```

```python
## drop date column 
df.drop('date',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['arr_time'].str.split(':').str[0]
df['Arrival_Minute'] = df['arr_time'].str.split(':').str[1]
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
df.drop('arr_time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Departure_hour'] = df['dep_time'].str.split(':').str[0]
df['Departure_Minute'] = df['dep_time'].str.split(':').str[1]
df['Departure_hour'] = df['Departure_hour'].astype(int)
df['Departure_Minute'] = df['Departure_Minute'].astype(int)
df.drop('dep_time',axis=1,inplace=True)
df.head()
```
